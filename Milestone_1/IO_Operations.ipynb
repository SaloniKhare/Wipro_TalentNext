#**MINI PROJECT**

```python
#PROJECT-1
import string

try:
    filename = input("Enter file name: ")

    file = open(filename, "r")

    lines = file.readlines()

    # Meeting Time
    line_count = len(lines)

    if line_count <= 12:
        print("Meeting time:", line_count, "AM")
    else:
        print("Meeting time:", line_count - 12, "PM")

    # Count word frequency
    freq = {}

    for line in lines:
        # Remove punctuation
        line = line.translate(str.maketrans("", "", string.punctuation))

        words = line.split()

        for word in words:
            word = word.capitalize()

            if word in freq:
                freq[word] += 1
            else:
                freq[word] = 1

    # Find most frequent word
    place = max(freq, key=freq.get)

    print("Meeting place:", place, "Street")

    file.close()

except FileNotFoundError:
    print("File not found.")

except Exception as e:
    print("Error:", e)
```

##**Hands on Assignments For I/O Operations**

```python
#1.
try:
    filename = input("Enter file name: ")

    file = open(filename, "r")

    print(file.read())

    file.close()

except FileNotFoundError:
    print("File not found.")
```

```python
#2.
try:
    filename = input("Enter file name: ")
    n = int(input("Enter number of lines: "))

    file = open(filename, "r")

    for i in range(n):
        line = file.readline()

        if line == "":
            break

        print(line, end="")

    file.close()

except FileNotFoundError:
    print("File not found.")

except ValueError:
    print("Invalid input.")
```

```python
#3.
try:
    filename = input("Enter file name: ")

    file = open(filename, "a")

    text = input("Enter text to append: ")

    file.write("\n" + text)

    file.close()

    print("Data appended successfully.")

except Exception as e:
    print("Error:", e)
```

```python
#4.
try:
    filename = input("Enter file name: ")

    file = open(filename, "r")

    lines = []

    for line in file:
        lines.append(line.strip())

    file.close()

    print(lines)

except FileNotFoundError:
    print("File not found.")
```

```python
#5.
try:
    filename = input("Enter file name: ")

    file = open(filename, "r")

    longest = ""

    for line in file:
        words = line.split()

        for word in words:
            if len(word) > len(longest):
                longest = word

    file.close()

    print("Longest word:", longest)

except FileNotFoundError:
    print("File not found.")
```

```python
#6.
try:
    filename = input("Enter file name: ")

    word = input("Enter word to search: ")

    file = open(filename, "r")

    count = 0

    for line in file:
        words = line.split()

        for w in words:
            if w.lower() == word.lower():
                count += 1

    file.close()

    print("Frequency of", word, "=", count)

except FileNotFoundError:
    print("File not found.")
```
