#**MINI PROJECT**

```python
#PROJECT-1

people = {
    "Jeff": "Is afraid of Dogs.",
    "David": "Plays the piano.",
    "Jason": "Can fly an airplane."
}

print("Original Dictionary:")
for name, fact in people.items():
    print(name + ":", fact)

# Change Jeff's fact
people["Jeff"] = "Is afraid of heights."

# Add a new person
people["Jill"] = "Can hula dance."

print("\nUpdated Dictionary:")
for name, fact in people.items():
    print(name + ":", fact)
```

```python
#PROJECT-2
n = int(input("Enter number of scores: "))

scores = []

for i in range(n):
    score = int(input("Enter score: "))
    scores.append(score)

scores = list(set(scores))   # Remove duplicate scores
scores.sort()                # Sort in ascending order

print("Runner-up score:", scores[-2])
```

```python
#PROJECT-3

students = {
    "Krishna": [67, 68, 69],
    "Arjun": [70, 98, 63],
    "Malika": [52, 56, 60]
}

name = input("Enter a name: ")

marks = students[name]
average = sum(marks) / len(marks)

print("Average percentage mark:", average)
```

```python
#PROJECT-4
text = input("Enter a string: ")

count = text.count("Alex")

print("Number of times Alex appears:", count)
```

###**STRING**

```python
#1.
string = input("Enter a string: ")

upper = 0
lower = 0

for ch in string:
    if ch.isupper():
        upper += 1
    elif ch.islower():
        lower += 1

print("Uppercase letters:", upper)
print("Lowercase letters:", lower)
```

```python
#2.
string = input("Enter a string: ")

if string == string[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
#3.
string = input("Enter a string: ")

result = string[:2] * len(string)

print(result)
```

```python
#4.
string = input("Enter a string: ")

if string.startswith("x"):
    string = string[1:]

if string.endswith("x"):
    string = string[:-1]

print(string)
```

```python
#5.
string = input("Enter a string: ")
n = int(input("Enter n: "))

last = string[-n:]

print(last * n)
```
